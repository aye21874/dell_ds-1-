```python
%matplotlib notebook
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.chdir("D:\Documents\dataset(dell)\DGA_ecommerce_casestudy")
df_or = pd.read_csv("orders_dataset.csv")
df_rv = pd.read_csv("order_reviews_dataset.csv")
df_oid = pd.read_csv("order_items_dataset.csv")
df_pro = pd.read_csv("products_dataset.csv")
df_tr = pd.read_csv("product_category_name_translation.csv")
```

```python
df_or = pd.merge(df_or,df_rv,on = 'order_id', how = 'inner')
df_pro = pd.merge(df_pro,df_tr,on = 'product_category_name', how = 'inner')
df_oid = pd.merge(df_oid,df_pro,on = 'product_id', how = 'inner')
df_or = pd.merge(df_or,df_oid,on = 'order_id', how = 'inner')

```

```python
df_re = df_or.filter(['product_category_name_english','review_score'])
```

```python
gk = df_re.groupby('product_category_name_english')
```

```python
stat = gk.describe()
```

```python
stat = stat['review_score']
```

```python
stat = stat.sort_values(by=['mean'])
```

```python
#taking product categories with average rating below 4 as prone to customer dissatisfaction
stat = stat[(stat['mean'] < 4)]
```
